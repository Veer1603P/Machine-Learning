```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([11,22,33,44,55])
y = np.array([10,9,32,31,54])
x_mean = np.mean(x)
y_mean = np.mean(y)
print("Array 1:",x)
print("Mean of array 1 :",x_mean)
print("Array 2:",y)
print("Mean of array 2:",y_mean)
slope= np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
print("Slope of the arrays :",slope)
intercept = y_mean - (slope*x_mean)
print("Intercept of the arrays is:",intercept)
y_pred = intercept + slope*x
print("Predicted values of y :",y_pred)
plt.plot(x,y)

```
