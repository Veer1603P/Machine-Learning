```python
import numpy as np
import pandas as pd

col_name=['Pregnant','Glucose','BP','ST','Insulin','BMI','DPF','Age','Outcome']
dataf=pd.read_csv("diabetes.csv" , header=None , names=col_name)
dataf=dataf.drop(0,axis=0)
dataf
```

```python
feature_cols=['Pregnant','Glucose','BP','Insulin','BMI','DPF','Age']
x = dataf[feature_cols]
y = dataf.Outcome
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train , y_test = train_test_split(x,y,test_size=0.25 , random_state=16)

#random mai after every 16 row , next row traininng ko jayega ; for variation testing  it is been used
#test_size woh test size declare karta h , 70 : 30 ; 75:25 ; 80:20 ;  /100 karke likhna h 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=16)
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
```
